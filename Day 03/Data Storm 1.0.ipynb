```python
gender = pd.get_dummies(df_train["Gender"],drop_first = True)
gender.columns = ['Male']
```

```python
marital_status = pd.get_dummies(df_train["MARITAL_STATUS"],drop_first = True)
marital_status
```

```python
#If single is 0 then marital_status is other
```

```python
age = pd.get_dummies(df_train['AGE'])
age.columns = ['Age 31-45','Age 46-65','Age Less than 30','Age More than 65']
age = age.drop(['Age Less than 30'],axis=1)
age
```

```python
#If Age 31-45, Age 46-65 and Age More than 65 all are 0 then Age is less than 30
```

```python
df_train = pd.concat([df_train,gender,education,marital_status,age],axis=1)
df_train
```
